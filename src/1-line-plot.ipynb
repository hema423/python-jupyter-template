### Exercise 1: Line Plot Customization

```python
import matplotlib.pyplot as plt
import numpy as np

# Define x values from 0 to 10
x = np.arange(0, 11)
y = x ** 2  # Square of x

# Create the plot
plt.plot(x, y, marker='o', linestyle='-', color='b', label="y = x²")

# Customize the plot
plt.title("Line Plot of y = x²")
plt.xlabel("X Values")
plt.ylabel("Y Values (Square of X)")
plt.grid(True)
plt.legend()

# Show the plot
plt.show()

```
