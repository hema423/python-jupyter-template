### Exercise 5: Subplot Layout

```python
import matplotlib.pyplot as plt
import numpy as np

# Create figure and subplots
fig, axes = plt.subplots(2, 2, figsize=(10, 8))

# Line chart (First subplot)
x = np.arange(0, 11)
y = x ** 2
axes[0, 0].plot(x, y, marker='o', linestyle='-', color='b', label="y = x")
axes[0, 0].set_title("Line Chart")
axes[0, 0].set_xlabel("X Values")
axes[0, 0].set_ylabel("Y Values")
axes[0, 0].grid(True)
axes[0, 0].legend()

# Bar chart (Second subplot)
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]
colors = ['blue', 'green', 'red', 'purple']
axes[0, 1].bar(categories, values, color=colors)
axes[0, 1].set_title("Bar Chart")
axes[0, 1].set_xlabel("Categories")
axes[0, 1].set_ylabel("Values")

# Scatter plot (Third subplot)
x_scatter = [1, 2, 3, 4, 5]
y_scatter = [2, 4, 6, 8, 10]
axes[1, 0].scatter(x_scatter, y_scatter, color='blue')
for i in range(len(x_scatter)):
    axes[1, 0].text(x_scatter[i], y_scatter[i], f'({x_scatter[i]}, {y_scatter[i]})', fontsize=10, ha='right')
axes[1, 0].set_title("Scatter Plot")
axes[1, 0].set_xlabel("X Values")
axes[1, 0].set_ylabel("Y Values")
axes[1, 0].grid(True)

# Pie chart (Fourth subplot)
labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 25, 20, 15]
axes[1, 1].pie(sizes, labels=labels, autopct='%1.1f%%', colors=['blue', 'green', 'red', 'purple'], startangle=140)
axes[1, 1].set_title("Pie Chart")

# Adjust layout
plt.tight_layout()

# Show plot
plt.show()

```
